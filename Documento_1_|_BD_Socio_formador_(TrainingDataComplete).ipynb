##**Analisis modelo de regresion lineal multiple**
Crear el mejor modelo de regression lineal multiple para cada variable cuantitativa: (Income, Age, Experience, CURRENT JOB_YRS, CURRENT_HOUSE_ YRS y Risk Flag), y comparar los coeficientes obtenidos en estos modelos con respecto a los coeficientes obtenidos en el mapa de calor.

```python
df4 = df.loc [: , ["Income","Age","Experience","CURRENT_JOB_YRS", "CURRENT_HOUSE_YRS", "Risk_Flag"]]
df4.head()
```

```python
Var_Indendientes = ['Income', 'Age', 'Experience', 'CURRENT_JOB_YRS', 'CURRENT_HOUSE_YRS', 'Risk_Flag']
Var_Dependientes = {'Income': 0, 'Age': 1, 'Experience': 2, 'CURRENT_JOB_YRS': 3, 'CURRENT_HOUSE_YRS': 4, 'Risk_Flag': 5}

listas_diferentes = []

for palabra in Var_Dependientes.keys():
    lista_diferente = [columna for columna in Var_Indendientes if columna != palabra]
    listas_diferentes.append(lista_diferente)

# Ahora listas_diferentes contendrá 6 listas diferentes, cada una eliminando una palabra del diccionario
for i, lista_resultante in enumerate(listas_diferentes):
    print(f"Lista {i + 1}: {lista_resultante}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
```

```python
modelos = {}
ecuaciones = {}
coef_corr = {}

for palabra in Var_Dependientes.keys():
    # Crear una nueva lista sin la palabra
    nueva_lista = [columna for columna in Var_Indendientes if columna != palabra]

    # Crear un DataFrame con las nuevas columnas
    df_nuevo = df4[nueva_lista]

    # Convertir el DataFrame en arrays numpy
    X = df_nuevo.to_numpy()
    y = df4[palabra].to_numpy()

    # Crear y ajustar el modelo de regresión lineal
    model = LinearRegression()
    model.fit(X, y)

    # Almacenar el modelo en el diccionario
    modelos[palabra] = model

    # Construir la ecuación de regresión
    coeficientes = model.coef_
    intercepcion = model.intercept_
    ecuacion = f"{palabra} = "
    for i, coef in enumerate(coeficientes):
        ecuacion += f"{coef:.2f}{'x'}"
        if i < len(coeficientes) - 1:
            ecuacion += " + "
    ecuacion += f" + {intercepcion:.2f}"
    ecuaciones[palabra] = ecuacion

    # Calcular el coeficiente de correlación (R cuadrado)
    y_pred = model.predict(X)
    coef_Deter=model.score(X,y)
    coef_Correl=np.sqrt(coef_Deter)
    coef_corr[palabra] = coef_Correl


for palabra, modelo in modelos.items():
    print(f"Modelo de regresión lineal multiple para '{palabra}':")
    print("Coeficientes:", modelo.coef_)
    print("Intercepción:", modelo.intercept_)
    print("Ecuación de regresión:", ecuaciones[palabra])
    print("Coeficiente de correlación:", coef_corr[palabra])
    print()
```
